```python
def training(X,Y):
    print("input = ",X)
    print("output = ",Y)
    eta=0.1
    print("eta= ",eta)
    size=X.shape[1]
    print("size=",size)
    w=np.random.randn(1,size)
    print(w)
    Etotal=1
    Erequired=0.1
    epo=0
    while(Etotal>Erequired or epo<150):
        Etotal=0
        epo=epo+1
        l=len(X)
        print(l)
        for i in range(l):
            x=X[i]
            print("x= ",x)
            print("w= ",w)
            d=Y[i]
            net=np.dot(w,x)
            print("net= ",net)
            print("d= ",d)
            E=0.5*(d-net)**2
            print("E= ",E)
            Etotal=Etotal+E
            print("Etotal= ",Etotal)
            delw=eta*(d-net)*x
            print("delw= ",delw)
            w=w+delw
            print("w= ",w)
            print("------------------------------------------------")
    print("FINAL w= ",w)
    print("epo= ",epo)
    return w
def testing(val,w):
    y=w[0][0]*val+w[0][1]
    print(y)
```

```python
X=np.array(([0.1,1],[0.2,1],[0.3,1],[0.4,1],[0.5,1],[0.6,1],[0.7,1]))
# Y=np.array([[49,63,58,60,58,61,60]])
# target=(49,63,58,60,58,61,60)
Y=(1.2,1.4,1.55,1.75,2.01,2.2,2.35)
w=training(X,Y)
print(w)
val=0.15
testing(val,w)
```

```python
def generate_data(n):
    x = np.linspace(0, 5, n)
    y = 1 / (1 + np.exp(-3 * x))
    return x, y
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_prime(x):
    return sigmoid(x) * (1 - sigmoid(x))
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
n = 51
X, y = generate_data(n)
w = np.random.randn()
b = np.random.randn()
learning_rate = 0.1
epochs = 1000
for epoch in range(epochs):
    y_pred = sigmoid(w * X + b)
    loss = mse_loss(y, y_pred)
    delw = np.mean(X * (y_pred - y) * sigmoid_prime(w * X + b))
    delb = np.mean((y_pred - y) * sigmoid_prime(w * X + b))
    w = w - learning_rate * delw
    b = b - learning_rate * delb
    if epoch % 100 == 0:
        print("Epoch {}: Loss = {:.4f}".format(epoch, loss))
plt.scatter(X, y, label="Target")
plt.plot(X, sigmoid(w * X + b), label="Sigmoid Neuron")
plt.legend()
```
